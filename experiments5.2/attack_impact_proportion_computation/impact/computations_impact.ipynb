```python
def create_df_impact_results_table(experiment,attacker_str, comb_num):
    
    list_results = []
    for exp in [experiment]:
        for attacker in [attacker_str]:
            for comb in [comb_num]:
                if attacker in ['atr']:
                    att_ = 'attr'
                else:
                    att_ = attacker
                dDICT = create_specific_dDICT_grid(exp, att_, comb_num = comb )
                if attacker in ['atr', 'dd']:
                    n_iters = 500
                elif attacker in ['rep','pd']:
                    n_iters = 100
                if exp ==2:
                    for key in dDICT:
                        dict_results = obtain_impact_info_exp2(dDICT[key], attacker, str(comb), num_iters = n_iters)
                        dict_results['solver_1'] = key
                        list_results.append(dict_results)
                if exp == 3:
                    for key in dDICT:
                        dict_results = obtain_impact_info_exp3(dDICT[key], attacker, str(comb), num_iters = n_iters)
                        dict_results['solver_1'] = key
                        list_results.append(dict_results)
    
    return pd.DataFrame(list_results)
```

```python
for e in [2,3]:
    for a in ['atr','rep','dd','pd']:
        for c in [1,2,3,4]:
            print(e,a,c)
            df = create_df_impact_results_table(experiment = e,attacker_str = a, comb_num = c)
            df.to_csv('impact_results_csv_files/experiment_'+str(e)+'_attacker_'+str(a)+'_comb_'+str(c)+'_.csv',index = False)
            print(e,a,c,'done!')
```
