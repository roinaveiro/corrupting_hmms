```python
import os
import pandas as pd
```

```python
def filter_best_comb_df(df, exp, solver, att):
    df1 = df[df['exp']==exp]
    df1 = df1[df1['solver']==solver]
    df1 = df1[df1['att']==att]
    return df1
```

```python
df = pd.read_csv('definitive_best_combination_loop_all_experiments.csv')
```

```python
filter_best_comb_df(df, 3 , 'MCTS', 'pd')
```
