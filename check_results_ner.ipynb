```python
import pickle
import os

import numpy as np
import pandas as pd
```

# Load Results

```python
exp = "w1_2.0_w2_1.0_sentence_41785"
```

```python
def load_exp(dirname, seconds):
    fnames = os.listdir(dirname)
    x = []
    for name in fnames:
        full_name = f'{dirname}{name}'
        with open(full_name, 'rb') as f:
            x.append(pickle.load(f))
    
    df = pd.DataFrame(x)
    df['seconds'] = seconds
    return(df)

def load_multiple_exp(list):

    df = load_exp(list[0][0], list[0][1])
    for i in list[1:]:
        tmp = load_exp(i[0], i[1])
        df = pd.concat([df,tmp])

    return(df)
```

```python
dirname1 = f"results/{exp}_1500/"
seconds1 = 1500
dirname2 =  f"results/{exp}_3000/"
seconds2 = 3000
dirname3 =  f"results/{exp}_6000/"
seconds3 = 6000
ll = [(dirname1, seconds1), (dirname2, seconds2), (dirname3, seconds3)]
df = load_multiple_exp(ll)
```

```python
df.to_csv(f"results/full_{exp}.csv", index=False)
```
